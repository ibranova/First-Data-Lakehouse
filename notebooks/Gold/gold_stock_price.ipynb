Data in the gold layer is typically aggregated, cleaned, and enriched, making it suitable for direct consumption by end users and downstream applications.

Create a star schema for the business to help analysts easily access and get the data for dashboards.

A star schema consists of a central fact table connected to multiple dimension tables. This structure simplifies queries and improves performance for analytics.

This schema enables analysts to quickly generate business insights and build dashboards by joining the fact table with relevant dimensions.

```python
query = """
SELECT ROW_NUMBER() OVER (PARTITION BY name ORDER BY date) AS id,
date, name, open, high, low, close, volume
FROM datalakehouse.sliver.silver_stock_prices
"""
df = spark.sql(query)
df.display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("datalakehouse.gold.fact_stock_prices")
```
